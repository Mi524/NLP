<p>2.Supervised Direction to get optimal k and b


```python
import pandas as pd 
import numpy as np 
import random
import matplotlib.pyplot as plt 

path = '/Users/tracy/Documents/PythonProjects/Competition/Titannic/titanic/train.csv'

content = pd.read_csv(path)
content.dropna(how='any',inplace=True)

age_with_fares = content[
 	(content['Age']>22) & (content['Fare']< 400)&(content['Fare']>130)
]
sub_fare = age_with_fares['Fare']
sub_age = age_with_fares['Age']

def func(age,k,b):
	return k * age + b

def loss(y,yhat):
	"""
	:param y: the real fares
	:param yhat: the estimated fares
	:return : the distance between real fare and estimated fare
	"""
	return np.mean(np.abs(y-yhat))

min_error_rate = float('inf')

losses = []
loop_times = 10000

change_directions = [
    (+1,-1),
    (+1,+1),
    (-1,+1),
    (-1,-1)   ]

best_k = random.random() * 20 -10
best_b = random.random() * 20 -10

def step(): return random.random() * 1 

while loop_times > 0:

	k_delta_direction, b_delta_direction = random.choice(change_directions )

	k_delta = k_delta_direction * step()
	b_delta = b_delta_direction * step()

	new_k = best_k + k_delta
	new_b = best_b + b_delta 

	estimated_fares = func(sub_age,new_k,new_b)
	error_rate = loss(y=sub_fare,yhat=estimated_fares)
	if error_rate < min_error_rate :
		min_error_rate = error_rate
		best_k,best_b = new_k,new_b
        
		direction = (k_delta_direction,b_delta_direction)
		#print('loop == {}'.format(loop_times))
		losses.append(min_error_rate)
		#print("f(age) = {} * age + {},with error rate:{}".format(best_k,best_b,min_error_rate))
	else:
		direction = direction = random.choice(list(set(change_directions) - {(k_delta_direction,b_delta_direction)}))
	loop_times -= 1
```

<p>3.Gradient Descent to get optimal k and b

```python
def func(age,k,b):
	return k * age + b

def loss(y,yhat):
	"""
	:param y: the real fares
	:param yhat: the estimated fares
	:return : the distance between real fare and estimated fare
	"""
	return np.mean(np.abs(y-yhat))

losses = []
loop_times = 10000

k_hat = random.random() * 20 -10
b_hat = random.random() * 20 -10

def derivate_k(y,yhat,x):
	 abs_values = [1 if (y_i - yhat_i) >0 else -1 for y_i,yhat_i in zip(y,yhat)]
	 return np.mean([a * -x_i for a, x_i in zip(abs_values,x)])

def derivate_b(y,yhat):
	abs_values = [1 if (y_i - yhat_i) >0 else -1 for y_i,yhat_i in zip(y,yhat)]
	return np.mean([a * -1 for a in abs_values])

learning_rate  = 1e-1

while loop_times > 0:
	k_delta = -1 * learning_rate * derivate_k(sub_fare,func(sub_age,k_hat,b_hat),sub_age)
	b_delta = -1 * learning_rate * derivate_b(sub_fare,func(sub_age,k_hat,b_hat))

	k_hat += k_delta 
	b_hat += b_delta 

	estimated_fares = func(sub_age,k_hat,b_hat)
	error_rate = loss(y=sub_fare,yhat=estimated_fares)

# 	print('loop == {}'.format(loop_times))
# 	print("f(age) = {} * age + {},with error rate:{}".format(k_hat,b_hat,error_rate))
	losses.append(error_rate)
	loop_times -= 1
```

```python
for station in station_dict.keys():
	for line in metro_dict.keys():
		#如果该站点处于某条线路中,提取出该站点左右两边的站
		metro_line_list =  metro_dict[line]
		if station in metro_line_list:
			station_index = metro_line_list.index(station)
			#如果是第一个
			if station_index == 0 :
				station_dict[station] = station_dict[station] + [metro_line_list[1]]
			elif station_index == len(metro_line_list) - 1:
				station_dict[station] = station_dict[station] + [metro_line_list[station_index-1]]
			else:
				station_dict[station] = station_dict[station] + [metro_line_list[station_index-1]]
				station_dict[station] = station_dict[station] + [metro_line_list[station_index+1]]


def get_sub_dict(dictionary,number):
	sub_dict = {}
	counter = 0
	for k,v in dictionary.items():
		counter += 1
		if counter == number : break
		else:
			sub_dict.update({k:v})
	return sub_dict

sub_dict = get_sub_dict(station_dict,20)

def is_goal(path,destination):
	# print(path[-1],destination)
	def _wrap(path,destination):
		if path[-1] == destination:
			return True
		else: return False
	return _wrap(path,destination)

def is_subset(list_long,list_short):
	short_length = len(list_short)
	subset_list = []
	for i in range(len(list_long)-short_length+1):
		subset_list.append(list_long[i:i+short_length])
	if list_short in subset_list:
		return True
	else: return False

def get_successors(start,destination,is_goal,connection_graph):
	pathes = [[start]]
	new_pathes = []

	while pathes:
		pathes_len = len(pathes)
		path = pathes.pop(0)
		frontier = path[-1]
		successors = connection_graph[frontier]
		for s in successors:
			if s in path:
				continue
			else:
				new_path = path + [s]
				# print(len(new_path))
				pathes.append(new_path)
				if is_goal(new_path,destination):
					new_pathes.append(new_path)
					#如果找到的线路全部路线超过站点的数量，停止
					if len(new_pathes) >= len(new_pathes[0]):
						pathes = []
		#加入条件退出，否则会无限地往下搜索
		if new_pathes:  #如果已经有了第一条线路,如果此时搜索的每条线路已经是第一条线路的两倍数量进行返回
			if len(path) >=  2 * len(new_pathes[0]):
				pathes = []
		# print('test')
	return new_pathes

#这个函数需要注意优先把站点归类到同一条线路，特别是第一个站点，按照排序循环发现有问题
#必须要取得匹配最长的列表，并且是从头开始匹配
def get_line_path(metro_dict,successor): 
	if not successor:
		return []
	else:
		line_record = {}
		for line in metro_dict.keys():   #地铁线路
			metro_list = metro_dict[line]
			for i in range(1,len(successor)+1):    #0,1,2,3
				sub_successor = successor[:i]  #[地铁线，地铁线]
				#如果该线路有这个站点
				if is_subset(metro_list,sub_successor) :
					#将线路和有重合的数字记录到字典
					line_record.update({line:i})	
				else: break  #如果一个站点都没出现 直接退出 i 循环，搜索下一条线路
		#获得最常的一条线路
		line_record = sorted(line_record.items(),key=lambda x:x[1],reverse=True)
		line = line_record[0]
		max_station = line[0]
		max_station_len = line[1]
		next_successor = successor[max_station_len:]
		line_path = [ max_station ] * max_station_len 
		return line_path + get_line_path(metro_dict,next_successor)


#把一段有重复连续地铁线的列表变成连续非重复的列表
def get_unique_line(line_path):
	line = []
	# print('lengthhhhhh',len(path))
	last = line_path[-1]
	while line_path:
		frontier = line_path.pop(0)
		if line_path and frontier != line_path[0]:
			line.append(frontier)
	line = line + [last]
	return line
```

```python
#加入必经路段 by_way
def search(successors,metro_dict,by_way,method):
	print('从"{}"到"{}"'.format(successors[0][0],successors[0][1]))
	by_way_str = '、'.join(by_way)
	print('附带条件需要经过 {}'.format(by_way_str))
	print('方案选择:',method)

	#加入by_way的部分,如果by_way是空集合不处理
	new_successors = []
	if by_way:
		for i in range(len(successors)):
			if is_subset(successors[i],by_way):
				new_successors.append(successors[i])

	if not new_successors:
		print('没有找到需要同时经过 {} 的路线'.format(by_way_str))
		new_successors = successors
	#判断经过多少条线路
	line_pathes = []

	def _combine_line_station(metro_dict,successor):
		# print('eeee',len(successor))
		combine_dict = {}
		line_path  = get_line_path(metro_dict,successor)

		for i in range(len(line_path)):
			if line_path[i] not in combine_dict.keys():
				combine_dict.update({line_path[i]:[successor[i]]})
			else:
				combine_dict[line_path[i]].append(successor[i])
		return combine_dict

	def _get_line_and_stations(metro_dict,successors):
		all_combine = []
		for successor in successors:
			combine_dict = _combine_line_station(metro_dict,successor)
			all_combine.append(combine_dict)
		return all_combine

	def _get_sorted_stations(line_station_combinations):
		sorted_list = sorted(line_station_combinations,key = lambda x: len(list(x.values())[0]))
		return sorted_list
	def _get_sorted_lines(line_station_combinations):
		sorted_list = sorted(line_station_combinations,key = lambda x: len(list(x.keys())))
		return sorted_list

	# print(successors)

	all_combine = _get_line_and_stations(metro_dict,new_successors)
	sorted_stations = _get_sorted_stations(all_combine)
	sorted_lines = _get_sorted_lines(all_combine)

	if method == '最短路径优先':
		return sorted_stations[0]
	elif method == '最小换乘优先':
		return sorted_lines[0]
	#从时间角度考虑，换乘假设5分钟，一个站点2分钟，计算需要的时间，取出最短的一个方案
	elif method == '综合优先':  
		duration = []
		#优先计算换乘少的
		#score = sorted(sorted_lines,key=lambda x: 5*len(list(x.keys())) + 2*len(list(x.values())[0]))
		duration_sort = sorted(sorted_lines,key=lambda x: len([list(x.keys())]*5 + [list(x.values())[0]]*2))
		return duration_sort[0]
```
